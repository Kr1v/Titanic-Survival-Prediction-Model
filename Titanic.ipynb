```python
import pandas as pd 
import numpy as np 
train_data=pd.read_csv('./train.csv')
test_data=pd.read_csv('./test.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
features=['Pclass','Sex','Age','SibSp','Parch']
X=pd.get_dummies(train_data[features])
X['Age']=X['Age'].fillna(X['Age'].median())
y=train_data['Survived']
train_X,val_X,train_y,val_y=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
acc_scores={}
for depth in [3,5,7,10,15]:
    for n in [25,50,100,200,250]:
        model=RandomForestClassifier(n_estimators=n,max_depth=depth,random_state=1)
        model.fit(train_X,train_y)
        predictions=model.predict(val_X)
        acc=accuracy_score(val_y,predictions)
        acc_scores[(n,depth)]=acc
perfect_combo=max(acc_scores,key=acc_scores.get)
print(perfect_combo)

```
